```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Simulated A/B Test Data
np.random.seed(42)
group_a = np.random.binomial(1, 0.20, 1000)  # Generic subject — 20% open rate
group_b = np.random.binomial(1, 0.27, 1000)  # Personalised subject — 27% open rate

# T-Test
t_stat, p_value = stats.ttest_ind(group_a, group_b)

print(f"Group A (Generic) Open Rate: {group_a.mean():.2%}")
print(f"Group B (Personalised) Open Rate: {group_b.mean():.2%}")
print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

if p_value < 0.05:
    print("Result: Statistically significant — personalised subject lines perform better.")
else:
    print("Result: No significant difference detected.")

# Visualisation
plt.bar(['Generic (A)', 'Personalised (B)'],
        [group_a.mean(), group_b.mean()],
        color=['#d9534f', '#5cb85c'])
plt.title('A/B Test — Email Open Rate Comparison')
plt.ylabel('Open Rate')
plt.savefig('ab_test_result.png')
plt.show()
```
